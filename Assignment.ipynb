```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```
data = pd.read_csv('/content/Laptop-Price.csv')
data
```

```
data.dtypes
```

```
#Central Tendency
numeric_df = data.select_dtypes(include=np.number)
print("Mean:\n", numeric_df.mean())
print("\nMedian:\n", numeric_df.median())
print("\nMode:\n", numeric_df.mode().iloc[0])
```

```
# Range
print("Range:\n", numeric_df.max() - numeric_df.min())

# Interquartile Range (IQR)
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
print("\nInterquartile Range (IQR):\n", IQR)

# Variance
print("\nVariance:\n", numeric_df.var())

# Standard Deviation
print("\nStandard Deviation:\n", numeric_df.std())
```

```
# Covariance
plt.figure(figsize=(8,4))
sns.heatmap(numeric_df.cov(), annot=True)
plt.title("Covariance Matrix for Dataset")
plt.show()
```

```
#Correlation
plt.figure(figsize=(8,4))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Matrix for Dataset")
plt.show()
```

we want to test whether SSD has a significant relationship with Price_euros.

    H₀: There is no relationship between SSD and company Price_euros (correlation = 0).
    H₁: There is a relationship between SSD and Price_euros (correlation ≠ 0).


Choose a Significance Level

We will use α = 0.05, which is a common significance level.



Perform the Pearson Correlation test

We have numeric data, we use a "Pearson correlation" test to evaluate the relationship between SSD and Price_euros.


```
# Perform Pearson correlation test

Price_euros = numeric_df['Price_euros']
SSD= numeric_df['SSD']

correlation, p_value = pearsonr(Price_euros,SSD)

print(f'Correlation: {correlation}')
print(f'P-value: {p_value}')
```

There is a moderate positive relationship between price and SSD capacity.

The correlation is highly significant (with the p-value effectively zero).
